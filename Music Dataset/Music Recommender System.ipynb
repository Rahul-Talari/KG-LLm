```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Music Dataset\Music dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
print(df.duplicated().sum())
```

```python
df.shape
```

```python
df.head()
```

```python
df['User-Rating']
```

```python
l=[]
for i in df['User-Rating']:
    l.append(i[:3])
df['User-Rating']=l
df
```

```python
df['Album/Movie']=df['Album/Movie'].str.replace(' ','')
df['Singer/Artists']=df['Singer/Artists'].str.replace(' ','')
df
```

```python
df['tags']=df['Singer/Artists']+' '+df['Genre']+' '+df['Album/Movie']+' '+df['User-Rating']
df['tags'][0]
```

```python
new_df=df[['Song-Name','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df
```
